```python
import pandas as pd
```

```python
new_york_tree_census_2015_dataset = "../dramane/new_york_tree_census_2015.csv"
new_york_tree_census_2015_df = pd.read_csv(new_york_tree_census_2015_dataset)
new_york_tree_census_2015_df.head()
```

```python
new_new_york_tree_census_2015_df = new_york_tree_census_2015_df[['tree_id', 'stump_diam', 'status', 'health', 'spc_latin', 'spc_common', 'address', 'zipcode', 'boroname', 'nta_name', 'latitude', 'longitude']]

new_new_york_tree_census_2015_df
```

```python
new_new_york_tree_census_2015_df['spc_latin_for_join']=new_new_york_tree_census_2015_df['spc_latin'].str.strip()

new_new_york_tree_census_2015_df['spc_latin_for_join']=new_new_york_tree_census_2015_df['spc_latin_for_join'].str.replace(" ",'')
new_new_york_tree_census_2015_df['spc_latin_for_join']=new_new_york_tree_census_2015_df['spc_latin_for_join'].str.lower()
new_new_york_tree_census_2015_df.sample(10)
```

```python
chars = "\`*_{}[]()>#+-.,!$:;%'&/?"
for c in chars:
    new_new_york_tree_census_2015_df['spc_latin_for_join'] = new_new_york_tree_census_2015_df['spc_latin_for_join'].str.replace(c,"")

new_new_york_tree_census_2015_df
```

```python
clean_new_york_tree_census_2015_df = new_new_york_tree_census_2015_df.drop_duplicates()
clean_new_york_tree_census_2015_df.head(20)
```

```python

clean_new_york_tree_census_2015_df.to_csv(r'../dramane/clean_new_york_tree_census_2015_data.csv', index = False)
```
