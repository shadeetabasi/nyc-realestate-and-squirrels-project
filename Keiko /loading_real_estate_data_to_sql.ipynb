```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#loaded CSV
real_estate ="data/nyc_real_estate_building_final_use_this.csv"
real_estate_df = pd.read_csv(real_estate)
real_estate_df.head()
```

```python
real_estate_df = real_estate_df.rename(columns={"Full_Address": "full_address",
                                               "BUILDING CLASS CATEGORY": "building_class_category",
                                               "SALE PRICE": "sale_price",
                                               "NEIGHBORHOOD": "neighborhood",
                                               "TAX CLASS AT PRESENT": "tax_class_at_present",
                                               "ZIP CODE": "zip_code",
                                               "BUILDING CLASS AT TIME OF SALE": "building_class_at_time_of_sale",
                                               "borough_name": "borough_name",
                                               "latitude": "latitude",
                                               "longitude": "longitude"})

real_estate_df.head()
```

```python
#connect engine
###note to group, update local host location and name of database to reflect what is on your system
connection_string = "postgres:postgres@localhost:5433/real_estate_and_trees"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#checking tables names 
engine.table_names()
```

```python
#loading data to postgresSQL
real_estate_df.to_sql(name='real_estate_data', con=engine, if_exists='append', index=False)
```
