```python
import pandas as pd
import numpy as np
import geojson
import json
from geojson import Feature, FeatureCollection, Point
```

```python
#import csv
real_estate_file = 'data/nyc_real_estate_building_final_use_this.csv'
real_estate_df = pd.read_csv(real_estate_file)


real_estate_df
```

```python
features = real_estate_df.apply(
    lambda row: Feature(geometry=Point((float(row['longitude']), float(row['latitude'])))),
    axis=1).tolist()

#all other columns as properties
properties = real_estate_df.drop(['latitude', 'longitude'], axis=1).to_dict('records')

#whole geojson object
feature_collection = FeatureCollection(features=features, properties=properties)
```

```python
with open('data/real_estate.geojson', 'w', encoding='utf-8') as f:
    json.dump(feature_collection, f, ensure_ascii=False)
```
