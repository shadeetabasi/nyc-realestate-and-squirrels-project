```python
import pandas as pd
import numpy as np
```

```python
#import csv
real_estate_file = 'data/nyc_real_estate_building_final_use_this.csv'
real_estate_df = pd.read_csv(real_estate_file)


real_estate_df
```

```python
#breaking out by borough

manhattan_df = real_estate_df.loc[real_estate_df['borough_name'] == 'Manhattan']

manhattan_bulding_df = pd.DataFrame(manhattan_df['building_class_category'].value_counts())

manhattan_bulding_df.reset_index(drop=False)

manhattan_bulding_df
```

```python
#breaking out by borough

queens_df = real_estate_df.loc[real_estate_df['borough_name'] == 'Queens']

queens_bulding_df = pd.DataFrame(queens_df['building_class_category'].value_counts())

queens_bulding_df.rename(columns={"Index": "Type of Building"})

queens_bulding_df
```

```python
#breaking out by borough

brooklyn_df = real_estate_df.loc[real_estate_df['borough_name'] == 'Brooklyn']

brooklyn_bulding_df = pd.DataFrame(brooklyn_df['building_class_category'].value_counts())

brooklyn_bulding_df.rename(columns={"Index": "Type of Building"})

brooklyn_bulding_df
```

```python
#breaking out by borough

bronx_df = real_estate_df.loc[real_estate_df['borough_name'] == 'Bronx']

bronx_bulding_df = pd.DataFrame(bronx_df['building_class_category'].value_counts())

bronx_bulding_df.rename(columns={"Index": "Type of Building"})

bronx_bulding_df
```

```python
#breaking out by borough

staten_island_df = real_estate_df.loc[real_estate_df['borough_name'] == 'Staten Island']

staten_island_bulding_df = pd.DataFrame(staten_island_df['building_class_category'].value_counts())

staten_island_bulding_df.rename(columns={"Index": "Type of Building"})

staten_island_bulding_df
```

```python
manhattan_bulding_df.to_csv('data/manhattan_building_class_data.csv', index = True)
```
